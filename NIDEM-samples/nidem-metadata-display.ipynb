```python
import json
import folium
import geopandas
```

# Loading the STAC JSON metadata file into Python

```python
metadata_pathname = "sample-output/NIDEM_25m.stac.json"
with open(metadata_pathname) as src:
    metadata = json.load(src)
```

# Print the contents of the JSON structure

```python
metadata
```

# Display the properties harvested and generated from the conversion process

```python
metadata["properties"]
```

# Display the vector information on an interactive map

```python
map_osm = folium.Map(location=[-39.727, 146.155], zoom_start=7)
map_osm.add_child(folium.features.GeoJson(metadata))
map_osm
```

# Loading the vector file using GeoPandas

```python
gdf = geopandas.read_file(metadata_pathname)
gdf.plot()
```
